# Loop

```python
for st in "BACE":
    print(st)
```

```python
for st in [1,2,3]:
    print(st)
```

```python
for k in {1:5,2:6,3:7}.keys():
    print(k)
```

```python
for v in {1:5,2:6,3:7}.values():
    print(v)
```

```python
for k,v in {1:5,2:6,3:7}.items():
    print(k,v)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(2,5):
    print(i)
```

```python
for i in range(5,2,-1):
    print(i)
```

```python
for i in range(20,40,3):
    print(i)
```

```python
count = 0
while count <3:
    print(count)
    count = count + 1
```
